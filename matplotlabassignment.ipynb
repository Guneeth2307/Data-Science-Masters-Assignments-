Matplotlib is a Python library used for creating various types of plots and visualizations. The `Pyplot` module within Matplotlib simplifies the process of creating and customizing plots. A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional Cartesian plane, where one axis represents one variable and the other axis represents another variable. Each dot on the plot corresponds to a single data point, and the position of the dot is determined by the values of the two variables for that data point.

Q3)

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0, 1, 2, 3, 4, 5])


y1 = np.array([0, 100, 200, 300, 400, 500])


y2 = np.array([50, 20, 40, 20, 60, 70])

y3 = np.array([10, 20, 30, 40, 50, 60])

y4 = np.array([200, 350, 250, 550, 450, 150])


plt.figure(figsize=(10, 8))


plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title('Line 1')
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green')
plt.title('Line 2')


plt.subplot(2, 2, 3)
plt.plot(x, y3, color='red')
plt.title('Line 3')


plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple')
plt.title('Line 4')


plt.tight_layout()

plt.show()
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide a figure into multiple smaller plots, rows, and columns, and then you can plot different data or visualizations in each of these subplots. This is useful for comparing and visualizing multiple datasets or aspects of data side by side in a single figure.

```python
import numpy as np
import matplotlib.pyplot as plt


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


data_to_plot = [box1, box2]

plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Value')
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of a data distribution that displays a summary of key statistical measures, allowing you to visualize the central tendency, spread, and potential outliers within a dataset. Box plots are particularly useful for comparing the distribution of data between different groups or datasets.
